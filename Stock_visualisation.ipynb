```python
import investpy
import pandas as pd
from matplotlib.pyplot import figure
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
df = investpy.get_stock_historical_data(stock     ='infy',
                                        country   ='india', 
                                        from_date ='01/01/2010',
                                        to_date   ='31/12/2020',
                                        order     ='ascending')

labels_to_remove = ['Open','High','Low','Volume','Currency']

df.drop(labels_to_remove, inplace=True, axis=1)

print(df)
# print(df.to_string()) for printing all the lines 

#plot Closing price graph for Infosys from Jan 2010 to Dec 2020
figure(figsize = (23,6), dpi =160)
plt.title("Infosys")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.plot(df)
```
